```python

import pandas as pd
a1 = ['송중기', '남', 1, 79, 93]
a2 = ['김나현', '여', 1, 85, 85]
a3 = ['권보아', '여', 1, 91, 78]
a4 = ['김범수', '남', 1, 97, 98]
a5 = ['이효리', '여', 2, 93, 79]
a6 = ['박효신', '남', 2, 78, 90]
a7 = ['한예슬', '여', 2, 87, 74]
a8 = ['민경훈', '남', 2, 99, 89]
col1 = ["", "성별", "반", "국어", "영어"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6, a7, a8], columns=col1).set_index("")

df1
```

```python
df1.groupby("성별")["영어"].agg("mean")
# by, data, 집계함수 3가지 조합으로 만들어짐
# 시리즈로 출력됨
```

```python
df1.groupby("성별", as_index=False)["영어"].agg("mean")
# as_index=False
# 하나의 열로 나오도록 옵션 추가
```

```python
df1.groupby("성별")["영어"].agg("mean").reset_index()
# reset_index()으로도 동일한 결과
```

```python
df1.groupby(["성별", "반"])[["국어", "영어"]].agg("mean")
# 국어, 영어는 배열안에 배열로 넣었음
# 하나의 열을 인덱싱할때 : df["열1"]
# 복수의 열을 인덱싱할때 : df[["열1", "열2", ....]]
```

```python
df1.groupby(["성별", "반"]).agg("mean")[["국어", "영어"]]
# 이렇게 써도 실행됨..
# 근본없는듯...
```

```python
df1.groupby(["성별", "반"]).agg("mean")
# data 대상을 넣지 않으면, 그룹바이 외 나머지 열이 모두 집계됨
```

```python
df1.groupby(["성별", "반"])[["국어", "영어"]].agg(["mean", "sum", "std", "max"])
# std : 표준편차
```

```python
#피벗테이블과 그룹바이는 결과만 본다면 별 차이가 없다?
# 그룹바이인 경우
df1.groupby(["성별", "반"])[["국어", "영어"]].agg("mean")
```

```python
# 피벗테이블인 경우
df1.pivot_table(["국어", "영어"], index="성별", columns="반", aggfunc="mean")

#columns 차이가 있음
```

```python
# columns를 index에 넣은 경우
df1.pivot_table(["국어", "영어"], index=["성별", "반"], aggfunc="mean")

# groupby는 columns가 없고, index에 모두 몰빵한 형태
# 집계용도로는 같음
```
