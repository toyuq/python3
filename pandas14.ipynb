```python

import pandas as pd

a1 = ['송중기', 1, 92, 82, 17]
a2 = ['권보아', 1, 89, 91, 17]
a3 = ['김나현', 1, 71, 91, 18]
a4 = ['박효신', 1, 90, 78, 18]
a5 = ['김선미', 1, 83, 78, 19]
a6 = ['강민경', 1, 81, 91, 19]
col = ['이름', '반', '영어', '국어', '나이']
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=col)
df2 = df1.copy()
df3 = df1.copy()
df4 = df1.copy()

df1
```

```python
df1.rank()
```

```python
df1["영어"].rank()
```

```python
df1["영어"].rank(ascending=False)
# ascending 기본값 true, 값이 작은게 1부터 시작
# false는 값이 높은게 1부터 시작
```

```python
df1["영어등수"] = df1["영어"].rank(ascending=False)
df1
```

```python
df2
```

```python
df2["평균"] = df2["국어"].rank(ascending=False)
df2

# method 기본값 average
# method = "average" 평균등수
```

```python
df2["min"] = df2["국어"].rank(ascending=False, method="min")
df2
# min 줄 수 있는 최저의 숫자 
# 등수 중복, 공동 작은 등수처리, 중복 수 만큼 등수 건너뜀
```

```python
df2["max"] = df2["국어"].rank(ascending=False, method="max")
df2
# max 줄 수 있는 높은 숫자 
# 등수 중복, 공동 큰 등수처리
```

```python
df2["first"] = df2["국어"].rank(ascending=False, method="first")
df2
# first 동점일 때 정렬한 열을 기준으로 우선순위 배분
# 등수 중복없음
```

```python
df2["dense"] = df2["국어"].rank(ascending=False, method="dense")
df2
# dense 등수 중복, 등수 순차처리
```
