```python
import pandas as pd
df1 = pd.DataFrame([["송중기", 92, 82], ["권보아", 89, 91], ["김나현", 71, 89], 
                    ["박효신", 90, 78]], columns=["이름", "영어", "국어"])
df2 = df1.copy()
df3 = df1.copy()
df4 = df1.copy()

df1
```

```python
df1["반"] = 1
df1 

# "반"이라는 열이 있으면 생성되고, 없으면 수정함 (덮어씌움)
```

```python
df1["반"] = 2
df1 
```

```python
df1["학점"] = ['A', 'B', 'C', 'D']
df1 
```

```python
df1["영어점수"] = df1["영어"]
df1 
```

```python
df1["평균"] = (df1["영어"] + df1["국어"]) /2
df1 
```

```python
# concat을 이용해서 열 순서 바꾸기
df2
```

```python
df2["반"] = 1
df2 
```

```python
df2.iloc[:, -1]
# ":" 행을 모두 가져옴
# "-1" 로하면 뒤에서 가져옴
```

```python
df2.iloc[:, 0:-1]
# 맨뒤에서 첫번째 빼고 나머지 선택(반 빼고)
```

```python
pd.concat([df2.iloc[:, -1], df2.iloc[:, 0:-1]], axis=1)
# axis 0은 행, 1은 열
```
