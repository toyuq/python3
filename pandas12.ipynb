```python
import pandas as pd
df1 = pd.DataFrame([["송중기", 92], ["권보아", 89], ["김나현", 71], ["박효신", 90]],
                    columns=["이름", "점수"])
df2 = pd.DataFrame([["A", 90], ["B", 80], ["C", 0]], columns=["학점", "점수"])
df3 = pd.DataFrame([["A", 100], ["B", 90], ["C", 80]], columns=["학점", "점수"])
df4 = pd.DataFrame([["A", 95], ["B", 85], ["C", 75]], columns=["학점", "점수"])
df5 = pd.DataFrame([[90, "lucky"], [80, "lucky"]], columns=["점수", "구분"])
t1 = ["2021-02-11", "2021-02-12", "2021-09-21", "2021-09-22", "2021-09-23"]
df6 = pd.DataFrame([t1], index=["날짜"]).T
df7 = pd.DataFrame([["2021-02-12", "설날"], ["2021-09-21", "추석"]], 
                   columns=["날짜", "내용"])
df6["날짜"] = pd.to_datetime(df6["날짜"])
df7["날짜"] = pd.to_datetime(df7["날짜"])
df8 = pd.DataFrame([["송중기", "국어", 99], ["송중기", "영어", 72], 
                    ["권보아", "국어", 92], ["권보아", "영어", 92]], 
                   columns=["이름", "과목", "점수"])
df9 = pd.DataFrame([[90, "A", "국어"], [0, "B", "국어"], [95, "A", "영어"], 
                    [0, "B", "영어"]], columns=["점수", "학점", "과목"])
```

```python
df1.sort_values("점수")

#merge_asof
# 1. 숫자나 날짜(시간)형 데이터만 가능함
# 2. 데이터셋이 정렬되어있어야한다
```

```python
df1 = df1.sort_values("점수")
df2 = df2.sort_values("점수")
df2
```

```python
pd.merge_asof(df1, df2)
# 근사값의 VLOOPUP
```

```python
pd.merge_asof(df1, df2, allow_exact_matches=False)
# 경계값 선택 옵션 (allow_exact_matches) (이상/초과 옵션)
# 기본값은 true 경계값 포함
# false는 경계값 미포함
# 예시) false 일시, 90점은
```

```python
df3 = df3.sort_values("점수")
df3
```

```python
pd.merge_asof(df1, df3)
# 김나현은 NAN이 됨
```

```python

# direction 기준 방향 옵션 (상한선 / 하한선 옵션)
# backword 하한
# forward 상한
# nearest 중간

pd.merge_asof(df1, df3, direction="forward")
# 박효신은 90점이지만, B학점받음
```

```python
pd.merge_asof(df1, df3, direction="forward", allow_exact_matches=False)
# 90점 미만부터 B학점 받게 옵션 수정
```

```python
df4 = df4.sort_values("점수")
df4
```

```python
pd.merge_asof(df1, df4, direction="nearest")
# 중간값 기준
# df4의 정렬로 인해 C,B,A 순서라서 90점은 B학점으로 처리됨
# df4의 정렬을 거꾸로 하면 오류가 나서 처리할 수 없음
# df4의 B학점을 85가 아니라 84.999로 처리해야 가능함
```

```python
df5 = df5.sort_values("점수")
df5
```

```python
pd.merge_asof(df1, df2)
```

```python
pd.merge_asof(pd.merge_asof(df1, df2), df5)
```

```python
pd.merge_asof(pd.merge_asof(df1, df2), df5, tolerance=2)

# 허용오차 
# tolerance=2
# 90, 91, 92 까지 포함
```

```python
df6 = df6.sort_values("날짜")
df7 = df7.sort_values("날짜")

pd.merge_asof(df6, df7, direction="nearest", tolerance=pd.Timedelta('2d'))
# 시계열 범위
# 앞뒤로 2일 (nearest, pd.Timedelta('2d'))
```

```python
pd.merge_asof(df6, df7, direction="backward", tolerance=pd.Timedelta('2d'))
# 시계열 범위
# 앞으로 2일 (nearest, pd.Timedelta('2d'))
```

```python
# on은 근사값 설정, by는 정확한 값 설정
pd.merge_asof(df8, df9, on="점수", by="과목")
```
